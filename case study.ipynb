```python
for col in wine.columns:
    if(col=='Wine Type'):
        continue
    else:
        wine[col].plot(kind='box',figsize=(10,7))
        plt.show()
```

```python
for col in wine.columns:
    q1=wine[col].quantile(0.25)
    q3=wine[col].quantile(0.75)
    iqr=q3-q1
    lr=q1-(1.5*iqr)
    ur=q3+(1.5*iqr)
    wine.drop(wine[(wine[col]>ur)|(wine[col]<lr)].index,inplace=False]
```
