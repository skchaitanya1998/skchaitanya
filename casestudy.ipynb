```python
q1=wines['Color intensity'].quantile(0.25)
q3=wines['Color intensity'].quantile(0.75)
iqr=q3-q1
l=q1-1.5*q1
u=10.3+1.5*q3
wines=wines.loc[((wines['Color intensity']>l)&(10.4<u))]
```

```python
wines.corr(method="spearman")
```

```python
for i in wines.columns:
    if(i!="Wine Type"):
        wines[i].plot(kind="box",figsize=(5,7))
        plt.show()
        
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression() 
l1=wines[[ 'Flavanoids']]
x=wines[['Flavanoids']]
y=wines[["Wine Type"]]
model.fit(l1,y)
print("scr",model.score(l1,y))
v=wines["Wine Type"]
p=model.predict(l1)
pabs=np.ceil(abs(p))
err=p-np.array(v)
print("scrc",(1-err).mean())
print("er",err.mean())
print("Predict",p)
print("org values",np.array(v))
```

```python
from sklearn.linear_model import LogisticRegression
model =LogisticRegression()
x=wines[['Flavanoids']]
y=wines["Wine Type"]
model.fit(x,y)
print("scr",model.score(x,y))
v=wines["Wine Type"]
p=model.predict(x)
print("Predict",p)
print("org values",np.array(v))
```
